```python
arbol(DATOS_BRUTOS)
```

```python
diccionario_excel = pd.ExcelFile(DATOS_BRUTOS / 'fd_enadid18.xlsx')
```

```python
diccionario_excel.sheet_names
```

```python
diccionario_migrante = diccionario_excel.parse("TMIGRANTE", skiprows=19)
```

```python
variables = diccionario_migrante[['Mnemónico', 'Nombre']].dropna(how = 'all')
```

```python
variables_dic = {}
for indice, valores in variables.iterrows():
    variables_dic[valores[0]] = valores[1]
```

```python
variables_dic
```

```python
datos_migrante.columns = [col.upper() for col in datos_migrante.columns]

datos_migrante.rename(columns = variables_dic, inplace = True)
```
