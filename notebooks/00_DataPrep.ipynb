```python
DATOS_MIGRANTE = Path(DATOS_BRUTOS / 'enadid18' / "conjunto_de_datos_tmigrante_enadid_2018")
arbol(DATOS_MIGRANTE)
```

```python
datos_migrante = pd.read_csv(DATOS_MIGRANTE / 'conjunto_de_datos/' / 'conjunto_de_datos_tmigrante_enadid_2018.csv')
```

```python
datos_migrante.head().T
```

```python
catalogos = DATOS_MIGRANTE.joinpath("catalogos").glob("*.csv")
# list(catalogos)
```

```python
datos_catalogos = pd.DataFrame()
for catalogo in catalogos:
    working_df = pd.read_csv(catalogo, encoding = 'latin1')
    working_df['pregunta'] = catalogo.stem
    datos_catalogos = pd.concat((datos_catalogos, working_df), ignore_index = True)
```

```python
datos_catalogos
```

```python
variables_dic = {}
for valor_pregunta in datos_catalogos['pregunta'].unique():
    working_df = datos_catalogos[datos_catalogos['pregunta'] == valor_pregunta]
    variables_dic[valor_pregunta] = {}
    for indice, valores in working_df[['cve', 'descrip']].iterrows():
        variables_dic[valor_pregunta][valores[0]] = valores[1]
```

```python
variables_dic
```

```python
variables_dic.keys()
```

```python
datos_migrante.columns
```

```python
for key in variables_dic.keys():
    if key in datos_migrante.columns:
        datos_migrante[key] = datos_migrante[key].map(variables_dic[key])
    else:
        pass
```

```python
datos_migrante.head().T
```

```python
arbol(DATOS_MIGRANTE)
```

```python
diccionario_datos = pd.read_csv(DATOS_MIGRANTE / 'diccionario_de_datos' / 'diccionario_datos_tmigrante_enadid_2018.csv', encoding = 'latin1')
```

```python
diccionario_datos.head(15)
```

```python
columnas_dic = {}
for indice, valores in diccionario_datos[['nombre_campo', 'nemónico']].iterrows():
    columnas_dic[valores[1]] = valores[0]
```

```python
columnas_dic
```

```python
datos_migrante.columns = datos_migrante.columns.str.upper().map(columnas_dic)
```

```python
datos_migrante.columns
```
